## Get data from dataset server api requests

```python
import requests
```

```python
base_url = "https://datasets-server.huggingface.co"
endpoint = "is-valid"
params = {"dataset" : "scientific_papers"}
```

```python
def query(endpoint, params):
    return requests.get(base_url + "/" + endpoint, params)
```

```python
response = query(endpoint, params)
```

```python
response.json()
```

```python
endpoint = "first-rows"
params = {
    "dataset": "scientific_papers",
    "config": "arxiv",
    "split": "train"
    }
response = query(endpoint=endpoint, params=params).json()
```
